## Loops:
#### What are loops:
A loop is a programming construct that allows you to execute a block of code repeatedly. Instead of writing the same code multiple times, you can use a loop to automate repetitive tasks.
#### Why we use loops:
1. Avoid code repetition
2. Process collections of data (lists, strings, etc.)
3. Execute code until a condition is met
4. Save time and make code more maintainable

#### Example without loop:


```python
print("Hello 1")
print("Hello 2")
print("Hello 3")
print("Hello 4")
print("Hello 5")
```

#### Example with loop:


```python
for i in range(1, 6):
    print(f"Hello {i}")
```

Both produce the same output, but the loop is cleaner and more flexible!

#### What are Iterations:
An iteration is a single execution of the loop body. Each time the loop runs through its code block, that's one iteration.
#### Example:

```python
count = 0
while count < 3:
    print(f"Iteration {count + 1}")
    count += 1
```

### Loop Control Statement:
Python provides  keywords to control loop execution:
1. Break statement
2. Continue Statement

#### break Statement
Meaning: Exit the loop immediately and stop completely.
Example:

```python
for i in range(1, 10):
    if i == 5:
        break  # Stop loop when i is 5
    print(i)

# Output: 1, 2, 3, 4
```

When i becomes 5, the loop stops. Numbers 5, 6, 7, 8, 9 never print.
Real use: Searching for something - once you find it, stop looking!

```python
fruits = ["apple", "banana", "mango"]
for fruit in fruits:
    if fruit == "banana":
        print("Found it!")
        break  # Stop searching
```

#### continue statement:
Meaning: Skip the current iteration and move to the next one.
Example:

```python
for i in range(1, 10):
    if i == 5:
        continue  # Skip when i is 5
    print(i)

# Output: 1, 2, 3, 4, 6, 7, 8, 9
```

Only 5 is skipped. Everything else prints normally.
Real use: Filtering - skip items you don't want!

```python
numbers = [1, -2, 3, -4, 5]
for num in numbers:
    if num < 0:
        continue  # Skip negative numbers
    print(num)

# Output: 1, 3, 5
```

Key Difference
break = Stop the ENTIRE loop 
continue = Skip THIS round only, keep going 

```python
# With break
for i in range(1, 6):
    if i == 3:
        break
    print(i)
# Output: 1, 2

# With continue
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
# Output: 1, 2, 4, 5
```

while true example :





```python
while True:
    choice = int(input("Enter your choice: "))
    if choice >= 1:
        print("hello")
    elif choice >= 2:
        print("byy")
    elif choice >= 3: 
        print("exiting")
    else:
        print("invalid choice")
```

What This Program Does:
This program runs in an infinite loop (because of while True) and keeps asking the user for input forever until you manually stop it with Ctrl+C.

##### The Problem:
The logic has a flaw! Only one condition will ever execute because of how elif works.

What Happens with Different Inputs:
##### Input: 4

Check: if choice >= 1 → TRUE (4 ≥ 1)
Prints: "hello"
Skips all elif blocks
Loop repeats

##### Input: 10

Check: if choice >= 1 → TRUE (10 ≥ 1)
Prints: "hello"
Skips all elif blocks
Loop repeats

##### Input: 0

Check: if choice >= 1 → FALSE (0 is not ≥ 1)
Check: elif choice >= 2 → FALSE (0 is not ≥ 2)
Check: elif choice >= 3 → FALSE (0 is not ≥ 3)
Goes to: else
Prints: "invalid choice"
Loop repeats

##### Input: -5

All conditions are FALSE
Prints: "invalid choice"
Loop repeats

#### Key Points:
1. while True creates an infinite loop 
2. it will keep running forever
3. No break statement - there's no way to exit the loop normally
4. The elif blocks are unreachable - since any number ≥ 1 will match the first condition, the second and third conditions (elif choice >= 2 and elif choice >= 3) will never execute



#### How to Fix It:
If you wanted different messages for different ranges, you should either:
##### Option 1: Use exact equality
#### Option 2: Add a break statement

```python
while True:
    choice = int(input("Enter your choice: "))
    if choice >= 1:
        print("hello")
        break
    elif choice >= 2:
        print("byy")
        break
    elif choice >= 3: 
        print("exiting")
        break
    else:
        print("invalid choice")
```
