# Assignment Operators 

## Q4. Write a program that assigns a value to a variable and then increases it by 5 using an assignment operator.

### Answer:

```python
a=5
print(f"a before using assignment operator = {a}")
a+=5 #adding with 5 by assignment operator
print(f"a after increasing value by assignment operator = {a}")
```

# Comparison Operators

## Q7.  Write a program to check whether two numbers are equal or not.

# Membership Operators

## Q13. Write a program to check if a number exists in a given list. 

### Answer:


```python
marks={60,45,68,21,34,56,29,7,36,39}
print(f"marks of students are: {marks}")
print("56" in marks) #checking if 56 is present in marks or not
print("56" not in marks) #checking if 56 is not present in marks or not
print("90" in marks) #checking if 90 is present in marks or not
print("90" not in marks) #checking if 90 is not present in marks or not
```

## Q14. Write a program to check if a word exists in a given string. 

### Answer:


```python
string={"hello","extra","snow","cell"}
print(f"words are: {string}")
print("hello" in string) #checking if hello is present in string or not
print("hello" not in string) #checking if hello is not present in string or not
print("world" in string) #checking if world is present in string or not
print("world" not in string) #checking if world is not present in string or not
```
