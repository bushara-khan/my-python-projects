#  Control Statements

In Python, control statements are used to control the flow of execution of a program — meaning they decide which code runs, when it runs, and how many times it runs.


### If-elif-else statement
It is a decision-making statement used to test multiple conditions in a program.

The if-elif-else statement is used to check multiple conditions one by one.
If the first condition is true, the if block runs. Otherwise, Python checks the elif conditions. If none are true, the else block runs.

### Nested if else
A nested if-else is an if-else statement written inside another if-else statement to check multiple conditions step by step.

Write a program to check if a number is less than 0. If not, then check whether it is equal to 0 or greater than 0.

### Match case statement:
 A match statement in Python is used to compare a single value against multiple possible cases and execute the block of code that matches.
It works like a switch-case statement in other programming languages and is useful for decision-making when there are multiple fixed options.

#### Key Points:
1. Introduced in Python 3.10.
2. Checks one variable against several cases.
3. If no case matches, a default case (_) can be used.
4. Makes code cleaner and more readable than multiple if-elif-else statements.

#### Syntax of Match Statement:


match variable:
    case value1:
        # code block if variable equals value1
    case value2:
        # code block if variable equals value2
    case value3:
        # code block if variable equals value3
    case _:  # default case
        # code block if no value matches



#### e.g. 1 
Write a program using match statement to check the day number (1–7) and print the day name (1=Monday, 2=Tuesday, …). If the number is not between 1–7, print “Invalid day”.

```python
print("Enter day number (1-7):")
day = int(input("Enter day number (1-7): "))

match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case 6:
        print("Saturday")
    case 7:
        print("Sunday")
    case _:
        print("Invalid day number. Please enter a number between 1 and 7.")
```

### e.g 2
Write a program using match statement to perform a simple calculator operation. Take input for operator (+, -, *, /) and two numbers, then perform the corresponding operation. If operator is invalid, print “Unknown operator”.

```python
print("Enter the first number:")
num1 = float(input("Enter the first number: "))
print("Enter the second number:")
num2 = float(input("Enter the second number: "))
print("Enter the operation (+, -, *, /):")
operator = input("Enter the operation (+, -, *, /): ")
match operator:
    case "+":
        result = num1 + num2
        print(f"The result of {num1} + {num2} is: {result}")
    case "-":
        result = num1 - num2
        print(f"The result of {num1} - {num2} is: {result}")
    case "*":
        result = num1 * num2
        print(f"The result of {num1} * {num2} is: {result}")
    case "/":
        if num2 != 0:
            result = num1 / num2
            print(f"The result of {num1} / {num2} is: {result}")
        else:
            print("Error: Division by zero is not allowed.")
    case _:
        print("Invalid operator. Please enter one of the following: +, -, *, /.")
```

#### e.g 4
Write a program using match statement to check student grade (A, B, C, D) and print the corresponding message (A=Excellent, B=Good, C=Average, D=Fail). If grade is not A–D, print “Invalid grade”.

```python
print("Enter the grade:")
grade = str(input("Enter the grade: "))

if grade in ["A", "B", "C", "D"]:

    match grade:
        case "A":
            print("Excellent")
        case "B":
            print("Good")
        case "C":
            print("Average")
        case "D":
            print("Fail")
else:    print("Invalid grade. Please enter a grade between A and D.")           
```

### Difference between if el and match:




# Match Statement vs If-Else Statement

| Feature / Basis      | If-Else Statement                                      | Match Statement                                         |
|---------------------|--------------------------------------------------------|--------------------------------------------------------|
| **Purpose**          | Check conditions and execute code if True            | Compare a single value against multiple cases        |
| **Condition Type**   | Any logical condition (`>`, `<`, `==`, `in`, etc.)   | Equality or pattern matching only                     |
| **Syntax**           | `if condition: ... elif condition: ... else: ...`    | `match variable: case value1: ... case _: ...`       |
| **Default Case**     | `else:`                                               | `case _:`                                             |
| **Readability**      | Can become long and nested for many conditions       | Cleaner and more readable for multiple fixed options |
| **Use Case**         | Flexible, can check ranges, expressions, complex logic | Best for fixed values or patterns (like switch-case) |
| **Evaluation**       | Evaluates conditions one by one                       | Variable is checked once against all cases           |
| **Nested Complexity**| Often needs nested ifs for multiple checks           | Less nesting required                                 |
