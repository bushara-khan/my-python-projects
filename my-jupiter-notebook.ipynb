# 9 Type Casting
What is type casting? Write a Python example to convert a string into an integer.

# Answer:
Type casting in Python means converting a value from one data type to another.
For example, converting a string to an integer, or an integer to a float.

```python
num_str = "25"          # This is a string
num_int = int(num_str)  # Convert string to integer

print(num_int)          # 25
print(type(num_int))    # <class 'int'>
```
