```python
L1 = [100, 900, 300, 400, 500]
START = 1
SUM = 0

for C in range(START, 4):
    SUM = SUM + L1[C]
    print(C, ":", SUM)
    SUM = SUM + L1[0] * 10

# Aapke mutabiq ye dono loop se bahar hain:

print(SUM)
```

```python
def fun(s):
    k = len(s)
    m = " "
    for i in range(0, k):
        if s[i].isupper():
            m = m + s[i].lower()
        elif s[i].isalpha():
            m = m + s[i].upper()
        else:
            m = m + 'bb'
    print(m) 
    print(len(m)) 

fun('schOol2@com')
```

```python
Data = ["P", 20, "R", 10, "S", 30]
Times = 0
Alpha = ""
Add = 0

for C in range(1, 6, 2):
    Times = Times + C
    Alpha = Alpha + Data[C-1] + "$"
    Add = Add + Data[C]

print(Times, Add, Alpha)
```

```python
Msg1 = "WeLcOME"
Msg2 = "GUeSTs"
Msg3 = ""

# Loop Msg2 ki length tak chal raha hai (yani 0 se 6 tak)
for I in range(0, len(Msg2)+1):
    if Msg1[I] >= "A" and Msg1[I] <= "M":
        Msg3 = Msg3 + Msg1[I]
    elif Msg1[I] >= "N" and Msg1[I] <= "Z":
        Msg3 = Msg3 + Msg2[I]
    else:
        Msg3 = Msg3 + "*"

print(Msg3)
```

```python
x = "hello world"

# Line 1: Slicing with start, end, and negative indices
print(x[:2], x[:-2], x[-2:])

# Line 2: Single index and a specific range
print(x[6], x[2:4])

# Line 3: Mixing positive start and negative end indices
print(x[2:-3], x[-4:-2])
```
